```python
%matplotlib inline
import fitsio as ft
import matplotlib.pyplot as plt
import numpy as np
```

```python
d = ft.FITS('/fs/ess/PHS0336/data/templates/SGA-2020.fits')
```

```python
d
```

```python
dd = d[1].read()
```

```python
dd.dtype.names
```

```python
dd.size
```

```python
'''
D26	float32	arcmin	Major axis diameter measured at the μ=26 mag arcsec−2
 r-band isophote based on SMA_SB26. If the r-band surface-brightness profile 
 could not be measured at this level, the diameter is set equal to 2.5×
 SMA_SB25 or 1.5×
 D25_LEDA, in that order of priority.
 
 
MAG_LEDA	float32	Vega mag	Approximate brightness (Note: this magnitude 
 estimate is heterogeneous in both bandpass and aperture but for most 
 galaxies it is measured in the B-band within D25_LEDA ; use with care.)
'''
```

```python
plt.axes(xlabel='RA', ylabel='sinDEC')
plt.scatter(dd['RA'][::50], np.sin(np.radians(dd['DEC'][::50])), dd['D26'][::50], 
            alpha=0.6, marker='.')
```

```python
plt.axes(xlabel='D26 [arcmin]', ylabel='MAG_LEDA', xscale='log', xlim=(0.1, 20))
plt.scatter(dd['D26'][::50], dd['MAG_LEDA'][::50], marker='.', alpha=0.2)
```
