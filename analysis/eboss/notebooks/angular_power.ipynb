```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('/home/mehdi/github/LSSutils')
from lssutils.dataviz import setup_color
```

```python
setup_color()
```

```python
def plot_cls(cap):    
    p = '/home/mehdi/data/eboss/data/v7_2/3.0/measurements/cl/'
    cl_now = np.load(f'{p}cl_{cap}_noweight_mainhighz_512_v7_2_main_512.npy', allow_pickle=True).item()['cl_gg']['cl']
    cl_nn = np.load(f'{p}cl_{cap}_known_mainhighz_512_v7_2_main_512.npy', allow_pickle=True).item()['cl_gg']['cl']
    cl_lin = np.load(f'{p}cl_{cap}_knownsystot_mainhighz_512_v7_2_main_512.npy', allow_pickle=True).item()['cl_gg']['cl']

    names = ['No Weight', 'Linear', 'Neural Net']
    for i, cl_i in enumerate([cl_now, cl_lin, cl_nn]):
        plt.plot(cl_i, label=names[i])

    plt.yscale('log')
    plt.xscale('log')
    plt.legend()
    plt.ylim(1.0e-7, 1.0e-3)
    plt.ylabel(r'C$_{\ell}$')
```

```python
plot_cls('NGC')
```

```python
plot_cls('SGC')
```
