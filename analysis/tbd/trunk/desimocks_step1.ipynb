```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import nbodykit.lab as nb
```

```python
from glob import glob
```

```python
files = glob('/home/mehdi/data/redshift0.9873/*ELG*DDsmu_rsd.json')
len(files)
```

```python
files[1].split('_DDsmu')[0].split('snap97_')[-1]
```

```python
for file_i in files:
    name = file_i.split('_DDsmu')[0].split('snap97_')[-1]
    data = nb.SimulationBox2PCF.load(file_i)
    r  = data.corr['r'][:,0]
    xi = data.corr['corr'][:,0]
    plt.plot(r, r**2*xi, label=name)
plt.ylabel(r'$r^{2}\xi$')
plt.xlabel('r [Mpc/h]')
plt.legend(bbox_to_anchor=(1.01, 1.0))
plt.title(r'$j_{\mu}=0$')
```
