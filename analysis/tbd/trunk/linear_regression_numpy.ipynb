```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = 3.0
b = -1
num = 100

noise = np.random.normal(0, 0.1, size=num)
x = np.random.uniform(size=num)
y = a*x+b + noise


plt.scatter(x, y, marker='.')
```

```python
class Model(object):
    
    def __init__(self):
        self.w = 0.0
        self.b = 1.0
    def __call__(self, x):
        return self.w * x + self.b
```

```python
linear = Model()
plt.scatter(x, y, marker='.')
plt.scatter(x, linear(x), marker='.', color='r')
```
